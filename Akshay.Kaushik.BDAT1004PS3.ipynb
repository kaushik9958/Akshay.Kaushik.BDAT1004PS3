Occupations 
Step 1. Import the necessary libraries 
Step 2. Import the dataset from this address. 
Step 3. Assign it to a variable called users 
Step 4. Discover what is the mean age per occupation 
Step 5. Discover the Male ratio per occupation and sort it from the most to the least 
Step 6. For each occupation, calculate the minimum and maximum ages 
Step 7. For each combination of occupation and sex, calculate the mean age 
Step 8. For each occupation present the percentage of women and men 

```python
import pandas as pd

# Import the dataset 
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
# The dataset uses a pipe '|' delimiter,
users = pd.read_csv(url, sep='|')
users.head()
```

```python
# Discover the mean age per occupation
mean_age_per_occupation = users.groupby('occupation')['age'].mean()
print("\nMean Age per Occupation:")
print(mean_age_per_occupation)
```

```python
# Find the Male ratio per occupation and sort it from most to least
male_ratio = users.groupby('occupation')['gender'].apply(lambda x: (x == 'M').sum() / len(x))
sorted_male_ratio = male_ratio.sort_values(ascending=False)
print("\nMale Ratio per Occupation (Sorted):")
print(sorted_male_ratio)

```

```python
# calculate the minimum and maximum ages
min_max_age_per_occupation = users.groupby('occupation')['age'].agg(['min', 'max'])
print("\nMinimum and Maximum Ages per Occupation:")
print(min_max_age_per_occupation)

```

```python
# calculate the mean age for each combination of occupation and gender, 
mean_age_by_occupation_and_gender = users.groupby(['occupation', 'gender'])['age'].mean()
print("\nMean Age by Occupation and Gender:")
print(mean_age_by_occupation_and_gender)

```

```python
# Calculate the percentage of women and men for each occupation
gender_percentage_per_occupation = users.groupby(['occupation', 'gender']).size() / users.groupby('occupation').size() * 100
print("\nGender Percentage per Occupation:")
print(gender_percentage_per_occupation)
```

### Import necessary libraries and Dataset

```python
import pandas as pd
import numpy as np

# Step 2: Import the dataset
file_path = 'wind.txt'
data = pd.read_csv(file_path, delim_whitespace=True)
data
```

```python
# Step 3: Replace the first 3 columns with a proper datetime index
data.columns = ['Year', 'Month', 'Day'] + list(data.columns[3:])
data['Date'] = pd.to_datetime(data[['Year', 'Month', 'Day']].astype(str).agg('-'.join, axis=1), errors='coerce')
data.set_index('Date', inplace=True)
data = data.drop(columns=['Year', 'Month', 'Day'])
data
```

```python
# Step 4: Fix the year 2061 issue
def fix_year(date):
    if date.year > 1980:
        return date.replace(year=date.year - 100)
    return date

data.index = data.index.map(fix_year)
data
```

```python
# Step 5: Ensure the index is of type datetime64[ns]
data.index = pd.to_datetime(data.index)

print("\n-------------------------------------------------------------------------\n")
# Step 6: Compute the number of missing values for each location
missing_values = data.isna().sum()
print("Missing values per location:\n", missing_values)

print("\n-------------------------------------------------------------------------\n")
# Step 7: Compute the total number of non-missing values
non_missing_total = data.notna().sum().sum()
print("Total non-missing values:", non_missing_total)

print("\n-------------------------------------------------------------------------\n")
# Step 8: Calculate the mean windspeed over all locations and times
overall_mean = data.stack().mean()
print("Overall mean windspeed:", overall_mean)

print("\n-------------------------------------------------------------------------\n")
# Step 9: Calculate location-wise statistics
loc_stats = data.describe().T[['min', 'max', 'mean', 'std']]
print("Location-wise statistics:\n", loc_stats)

print("\n-------------------------------------------------------------------------\n")
# Step 10: Calculate daily statistics
day_stats = data.aggregate(['min', 'max', 'mean', 'std'], axis=1)
print("Daily statistics:\n", day_stats.head())

print("\n-------------------------------------------------------------------------\n")
# Step 11: Average windspeed in January for each location
january_data = data[data.index.month == 1]
january_mean = january_data.mean()
print("Average windspeed in January for each location:\n", january_mean)
```

```python
print("\n-------------------------------------------------------------------------\n")
# Step 12: Downsample to a yearly frequency
yearly_data = data.resample('Y').mean()
print("Yearly data:\n", yearly_data.head())
```

```python
print("\n-------------------------------------------------------------------------\n")
# Step 13: Downsample to a monthly frequency
monthly_data = data.resample('M').mean()
print("Monthly data:\n", monthly_data.head())
```

```python
print("\n-------------------------------------------------------------------------\n")
# Step 14: Downsample to a weekly frequency
weekly_data = data.resample('W').mean()
print("Weekly data:\n", weekly_data.head())
```

```python
print("\n-------------------------------------------------------------------------\n")
# Step 15: Calculate weekly statistics for the first 52 weeks
weekly_stats = data.resample('W').aggregate(['min', 'max', 'mean', 'std']).iloc[:52]
print("Weekly statistics for the first 52 weeks:\n", weekly_stats)
```

## Question 6 
Create a line plot showing the number of marriages and divorces per capita in the 
U.S. between 1867 and 2014. Label both lines and show the legend. 
Don't forget to label your axes! 

```python
# Step 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Load the dataset
# Assuming the file is in the current working directory
file_path = 'us-marriages-divorces-1867-2014.csv'
data = pd.read_csv(file_path)

# Step 3: Inspect the dataset structure (Optional)
print(data.head())

# Step 4: Extract the required columns
# Assuming the dataset contains columns: 'Year', 'Marriages_per_1000', 'Divorces_per_1000'
years = data['Year']
marriages = data['Marriages_per_1000']
divorces = data['Divorces_per_1000']

# Step 5: Plot the data
plt.figure(figsize=(10, 6))

# Plot marriages per capita
plt.plot(years, marriages, label="Marriages per 1,000", color="blue", linestyle="-")

# Plot divorces per capita
plt.plot(years, divorces, label="Divorces per 1,000", color="red", linestyle="--")

# Add labels, title, and legend
plt.xlabel("Year", fontsize=12)
plt.ylabel("Number per 1,000 People", fontsize=12)
plt.title("Marriages and Divorces per 1,000 People in the U.S. (1867-2014)", fontsize=14)
plt.legend(loc="upper right")
plt.grid(True)

# Step 6: Show the plot
plt.tight_layout()
plt.show()
```

## Question 7 
Create a vertical bar chart comparing the number of marriages and divorces per 
capita in the U.S. between 1900, 1950, and 2000. 
Don't forget to label your axes! 

```python
# Step 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Load the dataset
file_path = 'us-marriages-divorces-1867-2014.csv'
data = pd.read_csv(file_path)

# Step 3: Filter data for the specified years
selected_years = [1900, 1950, 2000]
filtered_data = data[data['Year'].isin(selected_years)]

# Step 4: Extract data for the bar chart
years = filtered_data['Year']
marriages = filtered_data['Marriages_per_1000']
divorces = filtered_data['Divorces_per_1000']

# Step 5: Plot the bar chart
bar_width = 0.35  # Width of the bars
x_indexes = range(len(years))  

plt.figure(figsize=(8, 6))

# Create bars for marriages and divorces
plt.bar(x=[x - bar_width / 2 for x in x_indexes], height=marriages, width=bar_width, label='Marriages per 1,000', color='blue')
plt.bar(x=[x + bar_width / 2 for x in x_indexes], height=divorces, width=bar_width, label='Divorces per 1,000', color='red')

# Step 6: Add labels, title, and legend
plt.xticks(ticks=x_indexes, labels=years, fontsize=12)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Number per 1,000 People", fontsize=12)
plt.title("Marriages and Divorces per 1,000 People (1900, 1950, 2000)", fontsize=14)
plt.legend(loc="upper left")
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Step 7: Show the plot
plt.tight_layout()
plt.show()
```

## Question 8 
Create a horizontal bar chart that compares the deadliest actors in Hollywood. Sort 
the actors by their kill count and label each bar with the corresponding actor's name. 
Don't forget to label your axes! 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'actor_kill_counts.csv'  
data = pd.read_csv(file_path)

# Sort the data by 'Count' for better visualization
data = data.sort_values(by='Count')

# Plot
plt.figure(figsize=(12, 8))
plt.barh(data['Actor'], data['Count'], color='darkblue')

# Labels and title
plt.xlabel('Kill Count', fontsize=12)
plt.ylabel('Actor', fontsize=12)
plt.title('Deadliest Actors in Hollywood', fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Annotate each bar with the kill count
for index, value in enumerate(data['Count']):
    plt.text(value + 5, index, str(value), va='center', fontsize=10)

# Show the plot
plt.tight_layout()
plt.show()
```

## Question 9  
Create a pie chart showing the fraction of all Roman Emperors that were 
assassinated. 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'roman-emperor-reigns.csv'  
emperors_df = pd.read_csv(file_path)

# Analyze the Cause_of_Death column
# Classify emperors into "Assassinated" and "Other"
emperors_df['Assassinated'] = emperors_df['Cause_of_Death'].str.contains("assassinated", case=False, na=False)
assassinated_count = emperors_df['Assassinated'].sum()
total_count = len(emperors_df)

# Calculate fractions
data = [assassinated_count, total_count - assassinated_count]
labels = ['Assassinated', 'Other']

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(data, labels=labels, autopct='%1.1f%%', startangle=90, colors=['red', 'blue'])
plt.title('Fraction of Roman Emperors Assassinated')
plt.axis('equal') 
plt.show()
```

## Question 10 
Create a scatter plot showing the relationship between the total revenue earned by 
arcades and the number of Computer Science PhDs awarded in the U.S. between 
2000 and 2009. 
 
Don't forget to label your axes! 
Color each dot according to its year.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'arcade-revenue-vs-cs-doctorates.csv'  
data = pd.read_csv(file_path)

# Scatter plot
plt.figure(figsize=(10, 6))
scatter = plt.scatter(
    data['Total Arcade Revenue (billions)'],
    data['Computer Science Doctorates Awarded (US)'],
    c=data['Year'], 
    cmap='viridis',
    edgecolor='k'
)

# Add labels and title
plt.xlabel('Total Arcade Revenue (in billions USD)')
plt.ylabel('Computer Science Doctorates Awarded (US)')
plt.title('Relationship Between Arcade Revenue and CS Doctorates (2000-2009)')

# Add a colorbar to indicate the year
cbar = plt.colorbar(scatter)
cbar.set_label('Year')

# Display the plot
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```
